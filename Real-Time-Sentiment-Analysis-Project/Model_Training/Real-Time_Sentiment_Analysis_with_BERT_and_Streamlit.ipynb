```python
# Load and Test the Pre-Trained Sentiment Model

from transformers import AutoModelForSequenceClassification, AutoTokenizer, pipeline
import torch

# Load pre-trained tokenizer and model

model_name = "cardiffnlp/twitter-roberta-base-sentiment"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

# Create a sentiment analysis pipeline

sentiment_pipeline = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)

#Test with an example text
text = "I absolutely loved the service and product quality!"
result = sentiment_pipeline(text)

#print the result
print(result)
```

```python
# Convert Label IDs to Meaningful Sentiment
# Label mapping dictionary
labels = {
    'LABEL_0': 'Negative',
    'LABEL_1': 'Neutral',
    'LABEL_2': 'Positive'
}

# Sample prediction
text = "The product is not worth the money."
result = sentiment_pipeline(text)

# Extract and translate label
label_id = result[0]['label']
sentiment = labels[label_id]
confidence = result[0]['score']

print(f"Sentiment: {sentiment} (Confidence: {confidence:.2f})")
```
